### Exercice 3

Let $X$ be a random variable with probability density function $$f(x)=\frac{2}{\pi}\sqrt{1-x^2} \mathbf{1}_{(-1,1)}(x).$$
Note that $$f(x)\leq c g(x), \quad \text{where }g(x)=\frac{1}{2}\mathbf{1}_{(-1,1)}(x),\, c=\frac{4}{\pi}.$$


1. Generate $(X_1,\ldots,X_n)$ iid copies of $X$ using the acceptance-rejection method.

2. Draw the empirical distribution using an histogram and compare it with the probability density function $f.$

3. Provide a Monte Carlo estimator of $\mathbb{E}[X].$

### Solution

1. Complete the code.

```python
import numpy as np

n=10**4
X=np.zeros(n) #create a vector of n zeros

for i in range(n):
    x=(-1+2*np.random.rand(1)) #uniformly distributed on (-1,1) 
    u=np.random.rand(1)  #uniformly distributed on (0,1)
    while u>np.sqrt(1-x**2): #condition equivalent to c*u*g(x)>f(x) (check it!)
        x=(-1+2*np.random.rand(1))
        u=np.random.rand(1)
    X[i]=x
X
```

2. Complete the code 

```python
import matplotlib.pyplot as plt #library for plots

x=np.linspace(-1,1,n)#x-coordinate for density
y=(2/np.pi)*np.sqrt(1-x**2)#y-coordinate for density
    

plt.hist(X,bins=round(n**(1/3)),density=1,label="simulation")
plt.plot(x,y,"green",label='density')
plt.legend(loc='best')
```

3. Complete the code.

```python
y =np.mean(X)  #estimator
s =np.std(X,ddof=1) #standard deviation
print("Estimator:",y)
print("Condidence Interval 95%:",[y-1.96*s/np.sqrt(n),y+1.96*s/np.sqrt(n)])
print("Error:",abs(1.96*s/(np.sqrt(n))))
```
