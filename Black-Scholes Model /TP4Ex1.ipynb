### Exercise 1

Let $(W_t)_{t\in [0,T]}$ be a Brownian motion, $(S_t)_{t\in [0,T]}$ be given by $$ S_t := S_0 \text{exp}{\left(\left(r-\frac{\sigma^2}{2}\right)t + \sigma W_t\right)}$$ and $0=t_0<t_1<\cdots<t_m=T$ be a time grid such that $t_{i+1}-t_i=\frac{T}{m}$. For the numerical experiments, we take 
$$
r=0.1, \sigma=0.3, S_0=100,  T=1, m=100.
$$

1. Generate $(W_{t_1}, \ldots,W_{t_m})$ and draw a sample path $t\mapsto W_t$ of the Brownian motion (by linear interpolation). What happens when $m$ increases?

2. Generate $(S_{t_1}, \ldots,S_{t_m})$ and draw a sample path $t\mapsto S_t$ of the Black-Scholes model (by linear interpolation). What happens when you change the values of the parameters $r, \sigma, T$ ?

3. Generate $(\bar{S}^m_{t_1},\ldots,\bar{S}^m_{t_m})$ and draw a sample path $t\mapsto\bar{S}^m_{t}$ of the Euler discretization of $S$ (by linear interpolation). Compare it to the sample path of the Black-Scholes model from Question 2.

## Solution

1. 

```python
import numpy as np
import matplotlib.pyplot as plt #library for plots
from scipy.interpolate import interp1d

d=1
r,sigma,S0,T,m=0.1,0.3,100,10,500

times=np.linspace(0,T,m)

dt=times[1]-times[0]
dB=np.sqrt(dt)*np.random.randn(m-1)  #genera [d0,...,dm-2]: vettore che contiene m-1 valori di una gaussiana standard 
W0=np.zeros(1)


W=np.concatenate((W0,np.cumsum(dB)))

plt.plot(times,W)
```

2.

```python
S=S0*np.exp((r-sigma**2/2)*times+sigma*W)
plt.plot(times,S)
```

3.

```python
Se=S0*np.cumprod(1+r*dt+sigma*dB)


Se=np.concatenate(([S0],Se))


plt.plot(times,Se)
plt.plot(times,S)
```
